```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

***Creating a linear model for T2 and T6**

```python
temp_df = app_df[["T2","T6"]]
temp_df.shape
```

```python
temp_df["T2"]
```

```python
X = temp_df.iloc[:,0]
y = temp_df.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge,Lasso
X_train,X_test,y_train,y_test =train_test_split(X,y, test_size =0.2,random_state = 1)
```

Making the X_train to be in a (-1,1) array for compatibility with the model

```python
X_train = X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
```

```python
lmodel = LinearRegression()
linear_model = lmodel.fit(X_train,y_train)
pred_temp = linear_model.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
# calculating the r2 score and rounding it to 2dp.
score_r2 = r2_score(y_test,pred_temp)
score_r2.round(2)
```

```python
app_df.head()
```

```python
app_df.drop(["date","lights"],axis = 1 ,inplace =True)
```

```python
app_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
scaled_df = pd.DataFrame(scale.fit_transform(app_df),columns =app_df.columns)
scaled_df.head()
```

```python
Xm =scaled_df.iloc[:,1:]
ym = scaled_df.iloc[:,0]
Xm_train,Xm_test,ym_train,ym_test = train_test_split(Xm, ym, test_size = 0.3, random_state =42)
```

```python
model = lmodel.fit(Xm_train,ym_train)
y_pred = model.predict(Xm_test)
```

```python
MAE = mean_absolute_error(ym_test,y_pred).round(2)
RSS = ((ym_test - y_pred)**2).sum()
RSS = RSS.round(2)
```

```python
mse = mean_squared_error(ym_test,y_pred).round()
RMSE = np.sqrt(mse).round(3)
rscore = r2_score(ym_test,y_pred).round(2)
```

```python
print(f"mse:{mse}")
print(f"MAE:{MAE}")
print(f"RSS:{RSS}")
print(f"RMSE: {RMSE}" )
print(f"r^2:{rscore} " )
```

```python
coef_df = pd.DataFrame(list(Xm_train.columns)).copy()
coef_df.insert(len(coef_df.columns),"coefficients",model.coef_.transpose())
coef_df.sort_values(by="coefficients", ascending = False)
```

**Working with ridge and lasso regression**

```python
ridge_model = Ridge(alpha = 0.4)
Rmodel = ridge_model.fit(Xm_train,ym_train)
Rmodel_pred = ridge_model.predict(Xm_test)
```

```python
MSE_ridge = mean_squared_error(ym_test,Rmodel_pred)
print(f"MSE_ridge:{MSE_ridge}")
RMSE_ridge = np.sqrt(MSE_ridge)
print(f"RMSE:{RMSE_ridge }")
```

```python
lasso_model = Lasso(alpha =0.001)
lassomodel = lasso_model.fit(Xm_train,ym_train)
lassocoef_df = pd.DataFrame(list(Xm_train.columns)).copy()
lassocoef_df.insert(len(lassocoef_df.columns),"Lasso_coefficients",lassomodel.coef_.transpose())
lassocoef_df.sort_values(by="Lasso_coefficients", ascending = False)
```

```python
non_zero =[]
for i in lassocoef_df["Lasso_coefficients"]:
    if i != 0:
        non_zero.append(i)
```
